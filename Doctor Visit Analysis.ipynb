# Visualize the percentage of people getting government health insurance due to low income, due to old age and also the percentage of people having private health insurance

```python
#percentage of people getting government insurance due to low income
label=['yes','no']
Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to low income")
plt.show()
#percentage of people getting private insurance 
Y=df[df['private']=='yes']
N=df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting privat health insurance")
plt.show()
#percentage of people getting government insurance due to old age, disability or veteran status
Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt health insurance due to old age, disability or veteran status")
plt.show()

```

# Plot a horizontal bar chart to analyze the reduced days of  activity due to illness based on gender

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
#Creating the bar chart
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
#Adding the aesthetics
plt.title('Barchart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
#show the plot
plt.show()
```
